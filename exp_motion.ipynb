```python
import numpy as np

file_path = '/content/motion.txt'

array = np.genfromtxt(file_path, delimiter=' ', skip_footer=1, invalid_raise=False, filling_values=np.nan)

print(len(array))
```

```python
import matplotlib.pyplot as plt

number_of_struct = np.arange(1,25,1)

time  = np.arange(0, 4000, 1)
res = np.zeros(4000)

for i in range(25):
  if np.isin(i, number_of_struct):

    for j in range(4000):
        res[j] = array[j][i]

    plt.plot(time, res)
  else:
    pass
```

```python
pixel_values_dict = {}
names = ['10 sec', '25 sec', '35 sec', '45 sec']
name_number = 0

for i in range(0, 4):

  print(i)
  p_v = np.zeros(3)
  p_v[0] = array[4][i]
  p_v[1] = array[4][i+4]
  p_v[2] = array[4][i+8]
  print(i+4)
  print(i+8)

  pixel_values_dict[names[name_number]] = p_v
  name_number += 1

print(pixel_values_dict)
```
