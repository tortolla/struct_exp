```python
import numpy as np

file_path = '/content/calibration.txt'  # Замените на путь к вашему файлу

# Считывание файла и преобразование его содержимого в массив numpy
array = np.loadtxt(file_path)

print(array.shape)
```

```python
import matplotlib.pyplot as plt

time  = np.arange(0, 500, 1)
res = np.zeros(500)

number_of_struct = np.array([17, 14, 13, 23, 8, 16])

for i in range(25):
  if np.isin(i, number_of_struct):

    for j in range(500):
        res[j] = array[j][i]

    plt.plot(time, res)
  else:
    pass

```
